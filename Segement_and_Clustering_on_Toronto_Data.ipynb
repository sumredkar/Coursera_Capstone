```python
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
tables = pd.read_html(link,header=0)
df = tables[0]
df.head()
```

```python
# set the column names
df.columns = ['PostalCode', 'Borough', 'Neighborhood']


# apply condition as if borough is not assigned then ignore that row and if Neighthood is not assigned then add the value of Borough to the Neighborhood
for index, row in df.iterrows():
    if row['Neighborhood'] == 'Not assigned':
        row['Neighborhood'] = row['Borough']

    if row['Borough'] == 'Not assigned':
        df.drop([index], axis=0, inplace=True)

df.head()
```

```python
# print the shape of your dataframe
# reset dataframe index
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
# add 2 more columns the he dataframe
latitude = [] # create empty latitude list
longitude = [] # create empty logitude list

# function to get the latitude and logitude for each post code from csv file
def get_lat_lng_from_csv():
    location_coordinates = pd.read_csv('http://cocl.us/Geospatial_data')
    for index, row in df.iterrows():
        data = location_coordinates[location_coordinates['Postal Code'] == row['PostalCode']]
        latitude.append(float(data['Latitude']))
        longitude.append(float(data['Longitude']))
```

```python
# assign the list of latitude and logitude tot he dataframe

# call function to add data to the list
get_lat_lng_from_csv()

df['Latitude'] = latitude
df['Longitude'] = longitude
df.head()
```

```python
df
```

```python
# explore and create map for toronto

# get the latitude and logitude for toronto canada
address = "Toronto canada"
geolocator = Nominatim(user_agent='ex_explorer')
location = geolocator.geocode(address)
toronto_latitude = location.latitude
toronto_longitude = location.longitude
print("Toronto latitude: ", toronto_latitude)
print("Toronto longitude: ", toronto_longitude)
```

```python
# filter df in which borough contains toronto
df_toronto = df[df['Borough'].str.contains('Toronto',regex=False)]
df_toronto.reset_index(drop=True, inplace=True)
df_toronto
```

# Use K-Mean CLustering Algorithm

```python
k=5
toronto_clustering = df_toronto.drop(['PostalCode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans = KMeans(n_clusters = k, random_state=0).fit(toronto_clustering)
print(dir(kmeans))
print(len(kmeans.labels_))
```

```python
print(kmeans.labels_)
```

```python
# add kmeans label to the dataframe
df_toronto['kmeans_clustering_labels'] = kmeans.labels_
df_toronto
```

```python
# set the color scheme for clustering
color_arr = np.arange(k)
color_arr
```

```python
colors_array = cm.rainbow(np.linspace(0, 1, len(color_arr)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
rainbow
```

```python
cluster_map = folium.Map(location=[toronto_latitude,toronto_longitude], zoom_start=10)

for lat, lng, neighborhood, cluster in zip(df_toronto['Latitude'],df_toronto['Longitude'],df_toronto['Neighborhood'],df_toronto['kmeans_clustering_labels']):
    label = 'CLuster {}, {}'.format(cluster, neighborhood)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup=label,
        color=rainbow[cluster - 1],
        fill=True,
        fill_opacity=0.5,
        parse_html=False
    ).add_to(cluster_map)

cluster_map
```
